```python
import ROOT
 
from array  import array 
import numpy as np 
from paa01 import paaFile
```

```python

DataFile = paaFile("LA-mml-data_1h-180320-173703.paa" )

print( DataFile.paaGetTextHeader(0))
print( DataFile.paaGetTextHeader(1))
print( DataFile.paaGetTextHeader(2))
print( DataFile.paaGetTextHeader(3)) 
print( DataFile.paaGetTextHeader(4))

ps = DataFile.paaGetPulseSize()
pc = DataFile.paaGetPulseCount() 
tl = DataFile.paaGetThresholdLevel() 

print("Points per pulse: ", ps) 
print("Pulses in this file: ", pc)
print("threshold level: ", tl)
print() 
pulse_data_t = list(range(ps)) 
```

```python
tl = -600

triggs = [] 
ROOT.gDirectory.Delete("TBPH;1")

dtimes = ROOT.TH1F("TBPH", "Tiempo entre pulsos", 10, 0, 0,)  

for pulse_number in range(pc):
    pulse_data_v = DataFile.paaGetPulseRP(pulse_number) 
    for i in range(0,ps):
        if( pulse_data_v[i]< tl ):
            if(len(triggs ) ==0 ):
                triggs.append(i)
            else: 
                if ((i -triggs[-1]  )>20 ):
                    triggs.append(i)
    #print(j, triggs)

    if( len(triggs) > 1): 
        print(" Double peak found at pulse:" + str(pulse_number) + repr(triggs))
        #print(str(pulse_number)) 
        for k in range(1, len(triggs)):
            dtimes.Fill((triggs[k]-triggs[k-1])*0.008)
            print((triggs[k] -triggs[k-1])*0.008)

    triggs.clear()

fit = ROOT.TF1("FitFunction", "[0]*TMath::Exp(-x/[1]) + [2]", 0, 9)
fit.SetParNames("N", "t", "C") 
fit.SetParameters(300,2.2, 1)

dtimes.Fit("FitFunction", "R") 

canvas = ROOT.TCanvas(" A canvas", "Counts per channel", 800, 600 ) 
dtimes.Draw()
dtimes.SetXTitle("\\Delta t [\\mu s]")
dtimes.SetYTitle("Conteos")

canvas.Draw()
```

```python
 #--------------------------------------------------------------------------------------------------------------
#                            Procesamiento múltiple con umbral FIJO
#--------------------------------------------------------------------------------------------------------------
# Configuración global para todo el script
tl = -600   
time_per_sample = 0.008  # µs 
min_samples_between_triggers = 20  # Distancia mínima entre picos

# Limpiamos los histogramas previos
ROOT.gDirectory.Delete("TBPH;1")

# Crear histograma 
dtimes = ROOT.TH1F("TBPH", "Tiempo entre pulsos", 20, 0.75 , 10)

# Archivos a procesar que terminen en .paa
file_names = [
    "LA-mml-data_1h-180320-173703.paa",
    "LA-mml-data_1h-180320-174704.paa",
    "LA-mml-data_1h-180320-175704.paa",
    "LA-mml-data_1h-180320-180704.paa",
    "LA-mml-data_1h-180320-181704.paa",
    "LA-mml-data_1h-180320-182704.paa",
]

for fname in file_names:
    df = paaFile(fname)
    ps = df.paaGetPulseSize()   
    pc = df.paaGetPulseCount()   
    
     
    
    for pulse_number in range(pc):
        pulse = df.paaGetPulseRP(pulse_number)
        triggs = []
        
         
        for i, val in enumerate(pulse):
            if val < tl:   
                if not triggs or (i - triggs[-1]) > min_samples_between_triggers:
                    triggs.append(i)
        
        
        if len(triggs) > 1:
            for k in range(1, len(triggs)):
                delta_t = (triggs[k] - triggs[k-1]) * time_per_sample
                dtimes.Fill(delta_t)

# Ajuste exponencial
fit = ROOT.TF1("expo", "[0]*TMath::Exp(-x/[1]) + [2]", 0.2, 8)
fit.SetParNames("N", "tau", "C")
fit.SetParameters(100, 2.2, 1)

# Realizamos el ajuste y obtenemos los parámetros
fit_result = dtimes.Fit("expo", "RQ")
tau = fit.GetParameter(1)   
tau_err = fit.GetParError(1)   

# imprimimos los resultados
print(f"Muestreo total: {dtimes.GetEntries():.0f} intervalos")
print(f"Vida media τ = {tau:.5f} ± {tau_err:.3f} µs")

 
c = ROOT.TCanvas("c", "Ajuste de vida media", 800, 600)
dtimes.SetFillColor(ROOT.kAzure-9)
dtimes.SetLineColor(ROOT.kBlue+2)
dtimes.SetXTitle("\\Delta t [\\mu s]")
dtimes.SetYTitle("Conteos")
dtimes.GetXaxis().CenterTitle()
dtimes.GetYaxis().CenterTitle()
dtimes.Draw("hist")
dtimes.Draw("E same")  # Dibujar errores
c.Draw()
```
